```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std) # standard devation ploted
```

```python
sns.countplot(x='sex',data=tips) # plot total amles and females
```

```python
sns.countplot(x='time',data=tips)
```

**BOX PLOT**
Representaion of categorical data

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker') #comparing smoker and no smoker
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex') # comparing male and female
```

**VIOLINPLOT**

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True) # ek side male ek side female
```

**DISTPLOT**

```python
sns.distplot(tips['total_bill']) # distplot are univarent
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
# x-> total bill
# y->frequency
#means total bill how many times appears
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') # use in regression
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```
