**IMPORT LIBRARIES**

```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

**DATA**

```python
x=np.dot(np.random.random(400).reshape(200,2),np.random.random(4).reshape(2,2))
x.shape
#200 rows and 2 columns matrix col[0]=x,xol[1]=y random points
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
#above data is not centered, so lets centered at 0
```

```python
 x=preprocessing.scale(x)
 plt.scatter(x[:,0],x[:,1])
 plt.grid()
 # now we can see data is centered at 0 in below graph.
```

**PCA (Principal Component Analysis)**

```python
pca=PCA(n_components=2)
pca.fit(x)
```

```python
print(pca.components_)
#below matrix is eigen matrix
```

```python
print(pca.explained_variance_)
#below matrix is eigen values 
```

**LINALG SVD**

```python
#eigen values and eigen vector using linear algebra inbuilt lib
covar=np.cov(x.T)
w,v=linalg.eig(covar)
# w is eigen values
# v is eigen vectors
```

```python
w #eigen values
```

```python
v # eigen vector 
```

```python
u,s,vh=np.linalg.svd(covar)
```

```python
u
```

```python
s
```

**PLOT**

```python
X = np.zeros((2,1))
Y = np.zeros((2,1))
```

```python
plt.scatter(x[:,0],x[:,1],alpha=0.5)
plt.quiver(X, Y, u[:,0], u[:,1], scale=3) # quiver create arrow
plt.quiver(X, Y, v.T[:,0], v.T[:,1], scale=4, color='red')
plt.quiver(X, Y, pca.components_[:,0], pca.components_[:,1], scale=3, color='b')
plt.grid()
```

```python
# Directions of spreads is defined by the arrows
# here we can see spred of arrows defined correct
# eigen vectors in given scatter plot
```
