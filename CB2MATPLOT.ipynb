```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(10)
print(x)
y1=x**2
y2=2*x+3
print(y1)
print(y2)

plt.plot(x,y1)
plt.show()
plt.plot(x,y2)
plt.show()
```

**LINE PLOT**


```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.style.use('seaborn')
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.style.use('seaborn')
plt.plot(x,y1,color='red',label="Apple")
plt.plot(x,y2,color='green',label="kiwi")
plt.legend()
plt.xlabel("Time")
plt.ylabel("price")
plt.title("prices of Fruits")
plt.show()
```

**SCATTER PLOTS**

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
plt.scatter(x,y1,color='red',label='Apple',marker="o")
plt.scatter(x,y2,color='green',label='kiwi',marker="*")
plt.legend()
plt.xlabel("Time")
plt.ylabel("price")
plt.title("Price of fruits over time")
plt.figure(figsize=(5,5)) # adjust size of any plot
plt.show()
```
