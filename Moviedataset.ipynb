```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
freq_title={}
# key --> 5:1<-- value/ frequency
# key --> 8:2<-- value/frequency

for t in titles:
  length = len(t)
  
  try:
      freq_title[length]+=1
  except:
      freq_title[length]=1
print(freq_title)
 
```

```python
x1=np.array(list(freq_title.keys()))
y1=np.array(list(freq_title.values()))
plt.scatter(x1,y1)
plt.xlabel("Movies title per len")
plt.ylabel("No of movies")
plt.show()
```
